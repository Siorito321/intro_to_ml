## MAtplotlib example workflow

```python
#0. allows to show lpots inside jp notebook
%matplotlib inline
import matplotlib.pyplot as plt

#1.Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

#2.Setup plot
fig, ax = plt.subplots(figsize=(10,10))

#3.Plot the data
ax.plot(x, y)

#4.Customize plot
ax.set(title="Simple plot", xlabel="x-axis", ylabel="y-axis")

#5. Save and show
fig.savefig("../images/sample-plot.png")
```

# Two optins for subplots



```python
# Subplot option 1
fig, ((ax1, ax2),(ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

#plot to each different axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(100))
```

```python
# optoin two (get access to ax not by their names but by their position in grid)
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))
#PLot to each different index
ax[0,0].plot(x, x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1,1].hist(np.random.randn(100));
```

## Which version should i use: pyplot vs matplotlib OO method
* When plotting somethong quickly use pyplot method
* When plotting smth advanced use OO method

```python
over_50 = heart_desease[heart_desease["age"] > 50]
len(over_50), len(heart_desease)
```

```python
#pyplot method
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target", figsize=(10,6));
```

```python
# OO method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target", ax=ax);
ax.set_xlim([45, 100])
```

```python
#oo method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

#PLot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])

#Customize
ax.set(title="Heart desease and chol level",
      xlabel="Age", ylabel="Cholesterol")

#Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

#Add a horizontal line in the moddle (mean)
ax.axhline(over_50["chol"].mean(), linestyle="--");
```

```python
over_50.head()
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10), sharex=True)

#Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])

#Customize ax0
ax0.set(title="Chol and heart desies levels",
       ylabel="CCholesterol");

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

#Add a meanline
ax0.axhline(y=over_50["chol"].mean(), linestyle="--");

scatter1 = ax1.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"])
ax1.set(title="thalach plot", xlabel="age", ylabel="thalach")
ax1.legend(*scatter1.legend_elements(), title="Target")
ax1.axhline(y=over_50["thalach"].mean(), linestyle="--");

#Add a title to a figure
fig.suptitle("Heart desease Analysis", fontsize=16, fontweight="bold");
```

```python
#set the style
plt.style.use("seaborn-v0_8-whitegrid")

#oo method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

#PLot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap="winter") #changes the color scheem!

#Customize
ax.set(title="Heart desease and chol level",
      xlabel="Age", ylabel="Cholesterol")

#Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

#Add a horizontal line in the moddle (mean)
ax.axhline(over_50["chol"].mean(), linestyle="--", color="red");
```

```python
#customizing y and x
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1,
                               figsize=(10, 10), sharex=True) #they shaer same x axes

#Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                     cmap="winter")

#Customize ax0
ax0.set(title="Chol and heart desies levels",
       ylabel="CCholesterol");
# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

#Add a meanline
ax0.axhline(y=over_50["chol"].mean(), linestyle="--");

#change axes limits
ax0.set_xlim([50,80])
ax1.set_ylim([60,200])

scatter1 = ax1.scatter(x=over_50["age"],
                       y=over_50["thalach"],
                       c=over_50["target"],
                       cmap="winter")
ax1.set(title="thalach plot", xlabel="age", ylabel="thalach")
ax1.legend(*scatter1.legend_elements(), title="Target")
ax1.axhline(y=over_50["thalach"].mean(), linestyle="--");

#Add a title to a figure
fig.suptitle("Heart desease Analysis", fontsize=16, fontweight="bold");
```
